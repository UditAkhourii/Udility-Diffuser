# **Welcome to Udility Diffuser Playground**
### A Meta Llama-3.5 based labeled-image generation model.

Unlike any other image generation model, Udility Diffuser usage SVG scripting and general inference technology to generate illustrative images.

Current diffusion based models can not generate labelled data that can be utilised for illustrative and educational purpose. Udility Diffuser takes care of this by reverse engineering the image generation process through a text-based contexting process followed by a SVG creation tech.
